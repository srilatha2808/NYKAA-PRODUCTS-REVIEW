```python
lf=[]
for sent in data["review_text"]:
    l=[]
    ls=str(sent).lower()
    ls=re.sub("[^a-z]"," ",ls)
    for word in word_tokenize(ls):
        if word in sw:
            pass
        else:
            stems=PorterStemmer().stem(word)
            l.append(stems)
    lf.append(" ".join(l))
    
            
    
    
```

```python
lf
```

```python
data1=pd.DataFrame({"cleaned_data":lf})
```

```python
data1
```

```python
cv=CountVectorizer()
```

```python
tra_d=cv.fit_transform(data1["cleaned_data"])
```

```python
pd.DataFrame(tra_d.toarray(),columns=sorted(cv.vocabulary_.keys()))
```
